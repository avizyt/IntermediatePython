```python
import numpy as np
```

```python
# compute reciprocals
np.random.seed(0)

def compute_reciprocals(val):
    output = np.empty(len(val))
    for i in range(len(val)):
        output[i] = 1.0/ val[i]
    return output
```

```python
val = np.random.randint(1,10, size=5)
compute_reciprocals(val)
```

```python
big_arr = np.random.randint(1,100, size=1000000)
```

```python
%timeit compute_reciprocals(big_arr)
```

```python
print(compute_reciprocals(val))
print(1.0/ val)
```

```python
%timeit (1.0/ big_arr)
```

```python
theta = np.linspace(0, np.pi, 100)
print(theta)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.Figure()
plt.plot(theta, np.sin(theta))
plt.plot(theta, np.cos(theta))
plt.plot(theta, np.tan(theta))
plt.show()
```

```python
exp_sin = np.exp(np.sin(theta))
exp_cos = np.exp(np.cos(theta))
exp_tan = np.exp(np.tan(theta))
```

```python
plt.plot(theta, exp_sin)
plt.plot(theta, np.sin(theta))
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(theta, exp_cos)
plt.plot(theta, np.cos(theta))
plt.xlabel('theta')
plt.ylabel('cos(theta)')
plt.xticks()
plt.legend(exp_cos)
plt.show()
```
