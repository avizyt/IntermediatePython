## Gaussian Process Regression

```python
from sklearn.gaussian_process import GaussianProcessRegressor

model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the gaussian process fit
gp = GaussianProcessRegressor(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE)
```

## Density  and Contour plot


```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
plt.show()
```

```python
plt.contour(X,Y,Z, 20, cmap='RdGy')
plt.show()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.show()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
plt.show()
```

```python
contours = plt.contour(X,Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.show()
```

## Two dimensional histogram

```python
mean = [0, 0]
cov = [[1, 1], [1,2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

## Handwritten digit data

```python
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)
```

```python
fig, ax = plt.subplot(8,8, figsize=(6,6))
for i , axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform((digits.data))
```

```python
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,c=digits.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)
```
